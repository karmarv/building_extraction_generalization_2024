## Visualize images in `IEEE Bigdata Cup 2024: Building extraction`

```python
import os, time
import fiftyone as fo
import fiftyone.zoo as foz
```

```python
# Path of the dataset formatted in COCO format
dataset_dir="coco"
! tree -L 2 $dataset_dir
```

#### Load the dataset, using tags to mark the samples in each split

```python
# Name of the dataset in session
name = "export_coco_beg2024" + str(time.time())
# The splits to load 
splits = ["train"] # "train", "val", 
# The type of the dataset being imported
dataset_type = fo.types.COCODetectionDataset  # for example
```

```python
# Import the dataset
dataset = fo.Dataset.from_dir(
    name = name,
    dataset_type=dataset_type,
    data_path=os.path.join(dataset_dir, splits[0]),
    labels_path=os.path.join(dataset_dir, splits[0], "{}.json".format(splits[0])),
    tags=splits[0]
)
```

```python
print(dataset.stats())
```

### Creates a session and opens the App in the output of the cell

```python
session = fo.launch_app(dataset)
```

```python
# Replace active App instance with screenshot so App state is viewable offline
session.freeze()
```
